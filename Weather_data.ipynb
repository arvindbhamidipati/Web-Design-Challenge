```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats.stats import pearsonr
from scipy import stats
```

```python
weather_data_path = "Resources/cities.csv"
weather_metadata = pd.read_csv(weather_data_path)
```

```python
weather_metadata
```

```python
#latitude by temp

fig1_data = weather_metadata[['Lat','Max Temp']]
fig1_data

x = weather_metadata[['Lat']]
y = weather_metadata[['Max Temp']]

plt.scatter(x, y, marker='o', color = "blue")

plt.title("Latitude vs. Max Temperature") 
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

plt.show()
```

```python
#Latitude by humidity

fig1_data = weather_metadata[['Lat','Humidity']]
fig1_data

x = weather_metadata[['Lat']]
y = weather_metadata[['Humidity']]

plt.scatter(x, y, marker='o', color = "blue")

plt.title("Latitude vs. Humidity") 
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")

plt.show()
```
